```python
import requests
res = requests.get('http://ggoreb.com/api/idol.jsp')
res.text  # 문자열 확인
idol = res.json()
idol
```

```python
for id in idol:
    print(id['agency'])  #파싱 parsing
```

```python
#agency의 종류가 몇개인가
agency_list=[]

for id in idol:
    if id['agency'] in agency_list: #존재한다면
        pass                        #아무것도 하지 않음
    else :                           #존재하지 않는다면
        agency_list.append( id['agency'])
agency_list
```
