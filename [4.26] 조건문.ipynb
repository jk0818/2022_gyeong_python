```python
#하나의 숫자를 입력받아 홀수인지 짝수인지 판단하는 함수 작성
def is_odd(a):
    if a % 2 == 0 :
        return "짝수"
    else : 
        return "홀수"
```

```python
#1이상의 정수 N을 입력받은 후 1~N까지의 합을 구하는 함수 작성

def total (a):
    t = 0
    for i in range(1, a+1):
        t += i
    return tot
```
