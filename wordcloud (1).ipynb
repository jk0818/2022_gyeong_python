```python
from wordcloud import WordCloud
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
import requests
res = requests.get('http://ggoreb.com/python/file/날개.txt')
res.encoding = None
text = res.text
```

```python
icon = Image.open('bird.png')
mask = Image.new("RGB", icon.size, (255,255,255))
mask.paste(icon, icon)
mask = np.array(mask)
```

```python
wordcloud = WordCloud(
  font_path='c:/windows/fonts/truetype/nanum/NanumGothicEco.ttf',
  background_color='white',
  mask=mask).generate(text)

plt.figure(figsize=(6, 6))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
